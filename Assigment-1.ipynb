```python
## Import functions
import pandas as pd  #import pandas which is used to create dataframes
from wordcloud import WordCloud, STOPWORDS #import word cloud and stopwords
from matplotlib import pyplot as plt #import matplotlib(used to plot images)
```

## <font color=blue >Data Description</font>
> __There are two csv files one contains data mining definitions and the other contain the artificial intelligence definitions. attributes are as follows:__
> 1. Category: defines which category the record belongs to i.e., data mining/ artificial intelligence <br>__Level of measurment__: Nominal <br>__Example__: Data Mining
> 2. Date: Date when the data is extracted from the website <br>__Level of measurment__: Interval <br>__Example__: 01/01/2020
> 3. Website Name: Contains Human readable website names ,<br>__Level of measurment__: Nominal<br>__Example__: Microsoft
> 4. URL: Contains the weblink from where the content is extracted<br>__Level of measurment__: Nominal <br>__Example__: https://www.ibm.com/cloud/learn/data-mining
> 5. Content: Contains the definition extracted from the website<br>__Level of measurment__: Nominal <br>__Example__: Data mining is process of finding anamolies and patterns.
***

## <font color=blue >Data Import and wrangling</font>
> 1. Data is read from the CSV file using pandas 
> 2. Content column is extracted and concatenated so that the string can be passed to the word cloud
***

```python
data_mining = pd.read_csv(r'C:\Users\jarug\Downloads\CS-5402\data_mining.csv',encoding="latin-1")
# read datamining.csv file and store the data frame in data_mining variable
data_mining.head()
```

```python
ai = pd.read_csv(r'C:\Users\jarug\Downloads\CS-5402\AI.csv',encoding="latin-1")
# read ai.csv file and store the data frame in ai variable
ai.head()
```

```python
data_mining_text = ''.join([text for text in data_mining['Content']])
# concatenate the text present in the content column in the data frame and assign the string value to a variable
ai_text = ''.join([text for text in ai['Content']])
# concatenate the text present in the content column in the data frame and assing the string value to a variable
print(ai_text[:100])
print(data_mining_text[:100])
```

## <font color=blue >Results</font>
> 1. Seperate word clouds are generated for both categories ie., data mining and artificial intelligence (AI)
> 2. The most frequent words in data mining cloud (other than data mining) are Information, large, procrss, pattern,knowledge.
> 3. The most frequent words in artificial intelligence cloud (other than artificial intelligence) are human, machine, computer,intelligent, machine 
***

```python
wordcloud = WordCloud(width = 800, height = 800,background_color ='white',stopwords = STOPWORDS,
                      min_font_size = 10).generate(data_mining_text)
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```

```python
wordcloud = WordCloud(width = 800, height = 800,background_color ='white',stopwords = STOPWORDS,
                      min_font_size = 10).generate(ai_text)
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
